```python
unique_values = {
    "Region": df["Region"].unique()[:10],
    "Country": df["Country"].unique()[:10],
    "State": df["State"].unique()[:10],
    "City": df["City"].unique()[:10],
}
```

```python
df = df.drop(columns=["State"])
```

```python
df.head()
```

```python
df = df.rename(columns={
    "AvgTemperature": "avg_temperature",
    "Region": "region",
    "Country": "country",
    "City": "city",
    "Month": "month",
    "Day": "day",
    "Year": "year"
})
```

```python
df.head()
```

```python
# Reorder columns (Date first)
df = df.reindex(columns=["year", "month", "day", "region", "country", "city", "avg_temperature"])
```

```python
# Step 4: Missing Value Handling
missing_before = df.isna().sum()

# Drop rows where all are NaN
df = df.dropna(how="all")
```

```python
df["date"] = pd.to_datetime(
    df["year"].astype(str) + "-" + df["month"].astype(str) + "-" + df["day"].astype(str),
    errors="coerce"
)

```

```python
df.head()
```

```python
# Step 6: Format Standardization
for col in ["region", "country", "city"]:
    df[col] = df[col].astype(str).str.lower().str.strip()
```

```python

# Replace inconsistent country values (example: us/usa/united states → united states)
df["country"] = df["country"].replace({
    "us": "united states",
    "usa": "united states",
    "u.s.": "united states"
})
```

```python

# Final missing values after cleaning
missing_after = df.isna().sum()
```

```python
#save the cleaned ds
df.to_csv("cleaned_Daily Temperature.csv", index=False)
```
